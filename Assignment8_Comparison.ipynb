```
from sklearn.linear_model import LogisticRegressionfrom sklearn.svm import SVCfrom sklearn.ensemble import RandomForestClassifierX=[[1,2],[2,3],[3,4],[10,10]]y=[0,0,0,1]lr=LogisticRegression().fit(X,y)svm=SVC().fit(X,y)rf=RandomForestClassifier().fit(X,y)print(lr.score(X,y),svm.score(X,y),rf.score(X,y))
```
