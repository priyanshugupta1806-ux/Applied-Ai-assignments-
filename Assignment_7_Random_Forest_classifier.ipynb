```python
from sklearn.ensemble import RandomForestClassifier

# Sample data
X = [[1,2],[2,3],[3,4],[10,10],[11,11]]
y = [0,0,0,1,1]

# Create model
model = RandomForestClassifier(n_estimators=10)

# Train model
model.fit(X, y)

# Predict
predictions = model.predict(X)

print("Predictions:", predictions)
```
